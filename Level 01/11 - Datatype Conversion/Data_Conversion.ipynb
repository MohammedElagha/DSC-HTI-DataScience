# Data Conversion

refers to convert of datatype of a column.  

most common 3 types of Data Conversion:
- boolean to binary, and vice versa
- string categorial to number, and vice versa
- int to float, and vice versa

there are 2 methods of applying conversion:
- automatic (bool2int, int2bool, str2int, int2float, float2int)
- defined list of values (str2int, int2str)




---



##### Sample Data

```
cars = [['Car A', 4, 'Large', True, 0, '4'],
        ['Car B', 7, 'Very Large', False, 0, '4'],
        ['Car C', 4, 'Mid', False, 1, '4'],
        ['Car D', 2, 'Small', True, 1, '2']]
```

```
columns = ['name', 'capacity', 'size', 'is-sport', 'rear-wheel', 'doors']
```

```
import pandas as pd
```

```
df = pd.DataFrame(data=cars, columns=columns)
```

```
df
```



---



##### Boolean to Binary

in boolean to binary, no need to use manual/defined list, because the conversion is known (True -> 1 , False -> 0)

```
df['is-sport'] = df['is-sport'].astype(int)
```

```
df['is-sport']
```



---



##### Binary to Boolean

in boolean to binary, no need to use manual/defined list, because the conversion is known (!= 0 -> True , 0 -> False)

```
df['rear-wheel'] = df['rear-wheel'].astype(bool)
```

```
df['rear-wheel']
```



---



##### String to Number

###### Automatic
- string int to int
- string to int

string int to int

```
df['doors'] = pd.to_numeric(df['doors'], errors='ignore')
```

```
df['doors']
```

string to int

```
df['size'] = df['size'].astype('category').cat.codes
```

```
df['size']
```

###### Pre-defined List

```
df = pd.DataFrame(data=cars, columns=columns)
```

```
size = {'Small': 1, 'Mid': 2, 'Large': 3, 'Very Large': 4}
```

```
df['size'] = [size[item] for item in df['size']]
```

```
df['size']
```



---



##### Number to String

only pre-defined

```
capacity = {2: 'sport', 4: 'normal', 7: 'SUV'}
```

```
df['capacity'] = [capacity[item] for item in df['capacity']]
```

```
df['capacity']
```



---



Int ot Float, Float to Int

```
df['doors'] = df['doors'].astype(float)
df['doors']
```

```
df['doors'] = df['doors'].astype(int)
df['doors']
```
